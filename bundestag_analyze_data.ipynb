```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# mount drive to save data
from google.colab import drive
drive.mount('drive')
```

```python
# load data
linke = pd.read_csv("drive/My Drive/NLP_project/linke_origin_b.csv")
gruene = pd.read_csv("drive/My Drive/NLP_project/gruene_origin_b.csv")
spd = pd.read_csv("drive/My Drive/NLP_project/spd_origin_b.csv")
cdu = pd.read_csv("drive/My Drive/NLP_project/cdu_origin_b.csv")
fdp = pd.read_csv("drive/My Drive/NLP_project/fdp_origin_b.csv")
afd = pd.read_csv("drive/My Drive/NLP_project/afd_origin_b.csv")
```

```python
# stop word removal
nltk.download('stopwords')
german_stop_words = stopwords.words('german')

# define a function for stop word removal
def stop_word_removal(input):
    token = input.split()
    return ' '.join([w for w in token if not w in german_stop_words])

# before stop word removal, case normalization is performed
linke.Tweets = linke.Tweets.str.lower()
linke.Tweets = linke.Tweets.apply(stop_word_removal)

gruene.Tweets = gruene.Tweets.str.lower()
gruene.Tweets = gruene.Tweets.apply(stop_word_removal)

spd.Tweets = spd.Tweets.str.lower()
spd.Tweets = spd.Tweets.apply(stop_word_removal)

cdu.Tweets = cdu.Tweets.str.lower()
cdu.Tweets = cdu.Tweets.apply(stop_word_removal)

fdp.Tweets = fdp.Tweets.str.lower()
fdp.Tweets = fdp.Tweets.apply(stop_word_removal)

afd.Tweets = afd.Tweets.str.lower()
afd.Tweets = afd.Tweets.apply(stop_word_removal)
```

```python
# plot word clouds

parties=[linke, gruene, spd, cdu, fdp, afd]

for party in parties: 
  allWords = ' '.join( [twts for twts in party['Tweets']] )
  wordCloud = WordCloud(width = 500, height = 300, random_state = 21, max_font_size = 110).generate(allWords)
  
  plt.imshow(wordCloud, interpolation = "bilinear")
  plt.axis('off')
  plt.show()
```
